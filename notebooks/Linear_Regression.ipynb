```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def linear_regression(x_train, y_train):
    x_train = x_train.reshape(-1,1)
    x_train = np.insert(x_train, 0, 1, axis=1)
    theta = np.linalg.inv(x_train.T @ x_train) @ x_train.T @ y_train
    return theta
```

```python
def plot_regression_line(x_train, y_train, theta):
    plt.scatter(x_train, y_train, color='blue', label='Training Data')
    x_min = np.min(x_train)
    x_max = np.max(x_train)
    x_range = np.linspace(x_min, x_max, 100)
    x_range_bias = np.insert(x_range.reshape(-1, 1), 0, 1, axis=1)
    y_range = x_range_bias @ theta
    plt.plot(x_range, y_range, color='red', label='Regression Line')

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Linear Regression')
    plt.legend()
    plt.show()
```

```python
data = pd.read_csv('../data/Medical Price Dataset.csv')
```

```python
x_train = data[['age']].values
y_train = data['charges'].values
```

```python
theta = linear_regression(x_train, y_train)
print('Theta:', theta)
```

```python
plot_regression_line(x_train, y_train, theta)
```
